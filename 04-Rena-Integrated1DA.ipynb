```python
# Fill the missing values in 'year_of_release' with the most common value. 
data['year_of_release_mode'] = data.groupby(['name', 'platform'])['year_of_release'].transform(lambda x: x.fillna(x.mode()[0]) if len(x.mode()) == 1 else x.min())

# Check how many NaN there are. 
data['year_of_release_mode'].isnull().sum()
```

For the missing values that remain, extract the year in the game's name (e.g. '2005' in 'All-Star Baseball 2005') and fill them.

```python
# Make a copy of 'year_of_release', fill the empty values with zero, get the unique values, convert them into str, and
# save into a list. 
my_list = data['year_of_release'].fillna(0).unique().astype(int).astype(str).tolist()

# Remove the '0' element since it does not belong in 'year_of_release'.
my_list.remove('0')

# Verify it worked. 
#my_list

# Make a copy of 'year_of_release'.
name_copy = data['name']
data['name_with_year'] = name_copy

# Look for the year information in 'name'.
data['matched'] = data['name_with_year'].apply(lambda x: ' '.join([m for m in my_list if m in x])).replace('',np.nan)

# Make sure to restore 'year_of_release' at the right idex, only where a value is missing 
data['year_of_release_combined'] = data['year_of_release_mode'].fillna(data['matched'])

# Check how many NaN's there are. 
data['year_of_release_combined'].isnull().sum()
```

For the missing values that still remain, at first, I thought of grouping 'name' by sequels (e.g. "Rock Band" (year missing), "Rock Band 2" (2008), "Rock Band 3" (2010)), then subtracting 1 from the immediate predecessor (i.e. "Rock Band 2" is from '2008', so "Rock Band" will get '2007'). The minimum in 'year_of_release' is 1980. Let's fill the missing values with '1800'. This way, if we need to plot a graph including the data with no 'year_of_release' value, the x-axis will not be too wide. 


In all likelihood, we will not be needing the rows with missing 'year_of_release'. So instead of throwing away the data, work with a data slice that excludes this segment. 

```python
# Fill the missing values in 'year_of_release_mode'with '1000' to indicate the information is missing.
data['year_of_release_combined'] = data['year_of_release_combined'].fillna(1800)

# Check how many NaN there are. 
data['year_of_release_combined'].isnull().sum()
```

Assuming that the global total sales figures are normally distributed, we can find out the probability of a score occurring by standardising the scores ("z-scores"). "Standarding the scores" means creating the standard normal distribution by converting the group of data in our frequency distribution such that the mean is 0 and the standard deviation is 1. 

Z-scores are expressed in terms of standard deviations from their means, calculated as follows:

    z-score = (global sales total - mean) / standard deviation

These z-scores have a distribution with a mean of 0 and a standard deviation of 1. 

```python
# Write a function to draw a diverging bar plot for an easy comparison of z-scores. 
def plot_diverging_bars(df, total, y_grid, x_grid, colour_col, c_title, x_title, y_title):
    # Calculate the z-score of each platform. 
    df[x_grid] = (df[total] - df[total].mean()) / df[total].std()

    # Assign colours to z-scores. Show the negative figures in red, and the positive in green.
    df[colour_col] = ['red' if x < 0 else 'green' for x in df[x_grid]]

    # Plot a diverging bar plot to easily compare the z-scores. 
    sns.set_style("whitegrid")
    plt.figure(figsize = (8, 4))
    plt.hlines(y = df[y_grid], xmin = 0, xmax = df[x_grid], color = df[colour_col], alpha = 0.4, linewidth = 10)
    plt.suptitle(c_title, fontsize = 14)
    plt.ylabel(y_title, size='x-large', fontsize = 14)
    plt.xlabel(x_title, size= 'x-large', fontsize = 14)
    plt.show();

# Plot the diverging bar plot.
df_to_plot = sales_sum_pivot
x_axis = 'z_score'
y_axis = 'platform'
colour_column = 'colours'
sales_total = 'global_sales_total'
chart_title = "Z-Score Comparison between Platforms"
x_label = 'Z-Score Scale'
y_label = 'Platforms'

plot_diverging_bars(df_to_plot, sales_total, y_axis, x_axis, colour_column, chart_title, x_label, y_label)
```

```python
# Write a function to build a heatmap showing the changes in global sales totals across Years in each platform. 
def plot_heatmap(df, c_title, y_title, x_title):
    

    # Create a copy of the table where all the rows are shift by one downwards and compare twith the rows in the original table. 
    changes = (df - df.shift(+1)).transpose().fillna(0)

    # Build a heatmap with the shift. 
    plt.figure(figsize = (15, 9))
    sns.heatmap(changes, cmap = 'RdBu_r')
    plt.suptitle(c_title)
    plt.ylabel(y_title, size='x-large')
    plt.xlabel(x_title, size= 'x-large')
    plt.show();

# Plot the heatmap.
df_to_plot = sales_across_years_pivot
chart_title = "Global Sales Totals across Years"
y_label = 'Platforms'
x_label = 'Years'

plot_heatmap(df_to_plot, chart_title, y_label, x_label)
```

```python
# Create a data slice by filtering only PS3-related data.
ps3_data = data_2013_2016.query('platform == "PS3"')
```

```python
# Write a function to plot two scatterplots side by side.
def plot_two_scatters(df, x_grid1, x_grid2, c_y_grid, c_title, x_title1, x_title2, c_y_title):

    # Prepare the datasets.
    df1 = df[[x_grid1, c_y_grid]]
    df2 = df[[x_grid2, c_y_grid]]
    
    # Plot two scatterplots side by side: so a 1-row, 2-column figure overall. 
    # Plot the left scatterplot with a red line. 
    plt.figure(figsize=(10,6));
    plt.subplot(1, 2, 1)
    plt.scatter(df1[x_grid1], df[c_y_grid])
    plt.title(c_title)
    plt.xlabel(x_title1)
    plt.ylabel(c_y_title)
  
    # Plot the right scatterplot with a green line. 
    plt.subplot(1, 2, 2)
    plt.scatter(df2[x_grid2], df2[c_y_grid])
    plt.xlabel(x_title2)
      
    # space between the plots
    plt.tight_layout(4)

    # show plot
    plt.show()


# Plot scatterplots to compare the effects of user and critic scores on the global sales total. 
df_to_plot = ps3_data
common_y_axis = 'global_sales_total'
common_y_label = "Global Sales Total (USD Million)"
chart_title = "Impact of User and Professional Review Scores on PS3 Game Sales"
x_axis1 = 'user_score_filled'
x_label1 = "User Scores"
x_axis2 = 'critic_score_filled'
x_label2 = "Critic Scores"

plot_two_scatters(df_to_plot, x_axis1, x_axis2, common_y_axis, chart_title, x_label1, x_label2, common_y_label)
```

```python
# Calculate the correlation between the PS3 game sales totals and the user scores. 
print('Correlation coefficients between global sales total and review scores')
print('User Score:', ps3_data['global_sales_total'].corr(ps3_data['user_score_filled']))
print('Critic Score:', ps3_data['global_sales_total'].corr(ps3_data['critic_score_filled']))
```

There had been some not-bad-at-all sales despite the lack of user or critic reviews. However, in general, the higher the user scores, the higher the total sales figures. 

This pattern can be seen much more clearly in the 'critic_score' data, suggesting that professional reviewers' opinions have more weight than users'. 

With a correlation coeficient falling between 0 and 0.3, the user scores have a weak positive linear relationship with the total sales figures. Critic scores, on the other hand, show a slightly stronger linear relationship but its correlation coeficient still falls below 0.3. 

We can indeed see the difference in the scatterplots that, for the critic scores, the higher sales totals are more concentrated above the highest scores and less scattered as they are in the user scores scatterplot.

### Average user ratings of the Xbox One and PC platforms are the same. <a class="anchor" id="section_5_1"></a>

```python
# Prepare arrays only with user scores, one for Xbox One and another for PC.
xone_user_scores = no_null_user_scores.loc[no_null_user_scores['platform'] == 'XOne', 'user_score_filled']
pc_user_scores = no_null_user_scores.loc[no_null_user_scores['platform'] == 'PC', 'user_score_filled']

# View the datasets.
#xone_user_scores
#pc_user_scores
```

```python
# Write a function for peforming some preliminary checks in order to know which statistical test to choose.
def get_stats(name1, name2, df1, df2):
    
    # Check the sample sizes.
    sample_size1 = len(df1)
    sample_size2 = len(df2)
    print(name1, 'sample size:', sample_size1)
    print(name2, 'sample size:', sample_size2)
    print()

    # Calculate the means of the user scores.
    average_user_score1 = round(df1.mean(), 2)
    average_user_score2 = round(df2.mean(), 2)
    print(name1, 'average user score:', average_user_score1)
    print(name2, 'average user score:', average_user_score2)
    print()

    # Calculate the variances.
    variance1 = df1.var()
    variance2 = df2.var()
    print(name1, 'variance:', variance1)
    print(name2, 'variance:', variance2)
    print()

    # Calculate the standard deviation.
    std1 = df1.std()
    std2 = df2.std()
    print(name1, 'std:', std1)
    print(name2, 'std:', std2)

# Call the function.
sample_name1 = 'xone'
sample_name2 = 'pc'
sample1 = xone_user_scores
sample2 = pc_user_scores

get_stats(sample_name1, sample_name2, sample1, sample2)
```

The standard deviation figures of the two samples are rather close. Let's check their variances with Levene's test.

A null hypothesis takes the form of "the measures of location of two populations are equal to each other". So, in our case, the null hypothesis will be:

* H0 = "The variance of Xbox One's average user score is the same as that of PC's average user score."

In case the null hypothesis is rejected, i.e. there is sufficient evidence against it, we need to set up the hypothesis for the not-equal-to alternative. So, in our case, the alternative hypothesis will be:

* H1 = "The variance of Xbox One's average user score is different from that of PC's average user score."

Note: in Levene's test, a small p-value suggests that the populations do NOT have equal variances, i.e.

If pvalue < 0.05, it implies unequal variances, therefore, go for H1.
<br>If pvalue >= 0.05, it implies equal variances, therefore, go for H0.

Check the data distribution to decide whether to go for the median-based test or the mean-based test. 

Three variations of Levene’s test are possible. The possibilities and their recommended usages are:

```python
# Write a function that plot two boxplots side by side. 
def plot_side_by_side_boxplots(df1, df2, c_title, c_xlabel, c_ylabel):

    sns.set_style("whitegrid")
    plt.figure(figsize=(10,6))
    plt.suptitle(c_title)
    plt.xlabel(c_xlabel, size= 'x-large')
    plt.ylabel(c_ylabel, size='x-large')
    plt.ticklabel_format(useOffset = False, style = 'plain', axis='y')
    plt.boxplot([df1, df2])
    plt.show()
    
# Plot boxplots for the data slices. 
sample1 = xone_user_scores
sample2 = pc_user_scores
chart_title = "User Scores per Platform"
chart_xlabel = 'Platform (1 = Xbox One, 2 = PC)'
chart_ylabel = 'User Scores'

plot_side_by_side_boxplots(sample1, sample2, chart_title, chart_xlabel, chart_ylabel)
```

The graphs are skewed to the left. So go for the median-based test. 

```python
# Write a function that performs Levene's test.
def run_levenes_test(df1, df2, b, a):
    
    # Set the critical statistical significance. 
    alpha = a

    # Levene's test
    results = st.levene(df1, df2, center = b)
    print(b, '-based p-value:', results.pvalue)

    if results.pvalue < alpha:
        print('We reject the null hypothesis centered at the', b)
    else:
        print("We can't reject the null hypothesis centered at the", b) 
              
# Run the test. 
sample1 = xone_user_scores
sample2 = pc_user_scores
base = 'median'
critical_value = 0.05

run_levenes_test(sample1, sample2, base, critical_value)
```

The pvalue is smaller than 0.05, which means the variances of the two samples are not equal, i.e. we go for H1.

**Formulate the null and alternative hypotheses and explain how.**

We have two samples of average user scores: one of Xbox One, and another of PC. The two samples are independent of each other. We want to compare the mean user scores of these two samples. 

The null hypothesis will be:

* H0 = "Xbox One's average user score is equal to that of PC."

The alternative hypothesis will be:

* H1 = "Xbox One's average user score is different from that of PC."


**Explain the criterion you used to test the hypotheses and why.**

Use Welch's t-test
<br>Reason: Welch's t-test allows us to compare two data sets and not just the two mean values. The mean user scores of the Xbox One and PC samples are fairly close. Whether or not the difference between these numbers is significant depends on the variances of the samples, which were confirmed to be equal by Leven's test. The sizes of the two samples are very different.

Welch's test is an adaptation of Student's t-test, and is more reliable when the two samples have unequal variances and/or unequal sample sizes like they are in our case. 


When performing the statistical test, set the following criteria: 
* Set equal_var = False
<br>Reason: the variances of the two samples are deemed equal.
* Set the statistical significance level of 5%, i.e. alpha = 0.05
<br>Reason: since this is not a test that needs strict accuracy, we are using the most commonly used value. 

If the p-value is less than alpha, we reject the hypothesis. 

If pvalue < 0.05, we reject the hypothesis, therefore, go for H1.
If pvalue >= 0.05, we accept the hypothesis, therefore, go for H0.

```python
# Write a function that performs Welch's test.
def run_welchs_test(df1, df2, var_bool, a):

    # Take the standard accepted value for the critical statistical significance.
    alpha = a

    results = st.ttest_ind(df1, df2, equal_var=var_bool)

    print('p-value:', results.pvalue)

    if results.pvalue < alpha:
        print('We reject the null hypothesis')
    else:
        print("We can't reject the null hypothesis") 

# Run the test. 
sample1 = xone_user_scores
sample2 = pc_user_scores
var_equality = False
critical_value = 0.05

run_welchs_test(sample1, sample2, var_equality, critical_value)
```

The p-value tells us that there is a 17% probability of getting a difference that size or larger, assuming that H0 is correct. Since the p-value is larger than the critical value, it indicates that such an observed outcome would be very likely under the null hypothesis, meaning "Xbox One's average user score is equal to that of PC.". In other words, our hypothesis that the two average user scores are the same has been accepted.
